```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.japanesefoodieguide.nl/en/category/restaurants/'
r = requests.get(url)
r.text
```

```python
soup = BeautifulSoup(r.text)
print(soup.find('body').prettify())
```

```python
# https://www.crummy.com/software/BeautifulSoup/bs4/doc/#find-all
all_shops = soup.find_all(class_="shop")
for shop in all_shops:
    shop_anchor_tag = shop.find('a')
    shop_title, shop_link = shop_anchor_tag.contents[2].strip(), shop_anchor_tag['href']
    print('%s: %s' % (shop_title, shop_link))
```
